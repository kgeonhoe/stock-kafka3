```python
# daily_watchlist 테이블 구조 확인
con = duckdb.connect(database=db_path, read_only=True)

try:
    # 테이블 존재 확인
    tables = con.execute("SHOW TABLES").fetchall()
    print("데이터베이스 테이블들:", [table[0] for table in tables])
    
    # daily_watchlist 테이블이 있는지 확인
    if ('daily_watchlist',) in tables:
        # 테이블 구조 확인
        table_info = con.execute("PRAGMA table_info('daily_watchlist')").fetchall()
        print("\ndaily_watchlist 테이블 구조:")
        for col in table_info:
            print(f"  {col[1]} {col[2]} (NOT NULL: {col[3]}, PK: {col[5]})")
            
        # 데이터 확인
        count = con.execute("SELECT COUNT(*) FROM daily_watchlist").fetchall()[0][0]
        print(f"\ndaily_watchlist 레코드 수: {count}")
        
        if count > 0:
            sample = con.execute("SELECT * FROM daily_watchlist LIMIT 3").fetchall()
            print("샘플 데이터:", sample)
    else:
        print("\ndaily_watchlist 테이블이 존재하지 않습니다.")
        
except Exception as e:
    print(f"daily_watchlist 테이블 조회 오류: {e}")

con.close()
```

```python
# daily_watchlist 테이블을 자동 증가 ID로 재생성 (DuckDB 방식)
con = duckdb.connect(database=db_path, read_only=False)

try:
    # 기존 테이블 백업 (데이터가 있다면)
    try:
        backup_count = con.execute("SELECT COUNT(*) FROM daily_watchlist").fetchall()[0][0]
        print(f"기존 데이터 수: {backup_count}")
    except:
        backup_count = 0
        print("기존 테이블이 없거나 삭제됨")
    
    # 시퀀스 생성 (DuckDB의 자동 증가 방식)
    try:
        con.execute("DROP SEQUENCE IF EXISTS daily_watchlist_id_seq")
    except:
        pass
        
    con.execute("CREATE SEQUENCE daily_watchlist_id_seq START 1")
    print("시퀀스 생성 완료")
    
    # 기존 테이블 삭제
    con.execute("DROP TABLE IF EXISTS daily_watchlist")
    print("기존 테이블 삭제 완료")
    
    # 새 테이블 생성 (시퀀스 사용)
    con.execute("""
        CREATE TABLE daily_watchlist (
            id INTEGER PRIMARY KEY DEFAULT nextval('daily_watchlist_id_seq'),
            symbol VARCHAR NOT NULL,
            date DATE NOT NULL,
            condition_type VARCHAR NOT NULL,
            condition_value DECIMAL(10,4),
            market_cap_tier INTEGER,
            created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
            UNIQUE(symbol, date, condition_type)  -- 중복 방지
        )
    """)
    print("새 테이블 생성 완료 (자동 증가 ID 포함)")
    
    # 테스트 데이터 삽입 (ID 없이)
    con.execute("""
        INSERT INTO daily_watchlist (symbol, date, condition_type, condition_value, market_cap_tier)
        VALUES ('AAPL', '2025-07-25', 'bollinger_upper_touch', 0.985, 1)
    """)
    
    # 결과 확인
    result = con.execute("SELECT * FROM daily_watchlist").fetchall()
    print(f"테스트 삽입 결과: {result}")
    
    # 새 테이블 구조 확인
    table_info = con.execute("PRAGMA table_info('daily_watchlist')").fetchall()
    print("\n새 테이블 구조:")
    for col in table_info:
        print(f"  {col[1]} {col[2]} (NOT NULL: {col[3]}, PK: {col[5]})")
        
    print("\n✅ 테이블 재생성 완료! (자동 ID 생성)")
    
except Exception as e:
    print(f"❌ 테이블 재생성 실패: {e}")

con.close()
```

```python
# DuckDB 디렉토리 파일 상태 확인
import os
import glob
from datetime import datetime

duckdb_dir = '/home/grey1/stock-kafka3/data/duckdb'
airflow_duckdb_dir = '/data/duckdb'

print("🗂️ DuckDB 디렉토리 파일 목록 확인...")

# 로컬 디렉토리 확인
print(f"\n📁 로컬 디렉토리: {duckdb_dir}")
if os.path.exists(duckdb_dir):
    files = os.listdir(duckdb_dir)
    print(f"파일 수: {len(files)}")
    
    for file in sorted(files):
        file_path = os.path.join(duckdb_dir, file)
        if os.path.isfile(file_path):
            size_mb = os.path.getsize(file_path) / (1024 * 1024)
            mtime = datetime.fromtimestamp(os.path.getmtime(file_path))
            print(f"  📄 {file} - {size_mb:.2f} MB ({mtime.strftime('%Y-%m-%d %H:%M')})")
else:
    print("❌ 로컬 디렉토리가 존재하지 않습니다")

# Airflow 컨테이너 디렉토리 확인
print(f"\n🐳 Airflow 디렉토리: {airflow_duckdb_dir}")
if os.path.exists(airflow_duckdb_dir):
    files = os.listdir(airflow_duckdb_dir)
    print(f"파일 수: {len(files)}")
    
    for file in sorted(files):
        file_path = os.path.join(airflow_duckdb_dir, file)
        if os.path.isfile(file_path):
            size_mb = os.path.getsize(file_path) / (1024 * 1024)
            mtime = datetime.fromtimestamp(os.path.getmtime(file_path))
            print(f"  📄 {file} - {size_mb:.2f} MB ({mtime.strftime('%Y-%m-%d %H:%M')})")
else:
    print("❌ Airflow 디렉토리가 존재하지 않습니다")

# 백업 파일 패턴 찾기
print("\n🗃️ 백업 파일 분석...")
backup_patterns = [
    "*.backup*",
    "*.corrupted*", 
    "*_backup_*"
]

all_backup_files = []
for directory in [duckdb_dir, airflow_duckdb_dir]:
    if os.path.exists(directory):
        for pattern in backup_patterns:
            backup_files = glob.glob(os.path.join(directory, pattern))
            all_backup_files.extend(backup_files)

if all_backup_files:
    print(f"총 백업 파일 수: {len(all_backup_files)}")
    total_size = 0
    for backup_file in sorted(all_backup_files):
        size_mb = os.path.getsize(backup_file) / (1024 * 1024)
        total_size += size_mb
        mtime = datetime.fromtimestamp(os.path.getmtime(backup_file))
        print(f"  🗃️ {os.path.basename(backup_file)} - {size_mb:.2f} MB ({mtime.strftime('%Y-%m-%d %H:%M')})")
    
    print(f"\n💾 총 백업 파일 크기: {total_size:.2f} MB")
else:
    print("백업 파일이 없습니다.")
```

```python
# 데이터베이스 복구 및 백업 파일 정리
import shutil
import os
from datetime import datetime

print("🔧 데이터베이스 복구 작업 시작...")

# 1. 가장 큰 백업 파일 (데이터가 있는 파일) 찾기
backup_files = [
    ('/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_044639', 3.76),
    ('/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_140005', 2.51),
    ('/home/grey1/stock-kafka3/data/duckdb/stock_data_replica.db', 2.51)
]

# 가장 큰 파일을 메인 DB로 복구
source_file = backup_files[0][0]  # 3.76MB 파일
main_db_path = '/home/grey1/stock-kafka3/data/duckdb/stock_data.db'
airflow_db_path = '/data/duckdb/stock_data.db'

print(f"📦 복구 소스: {os.path.basename(source_file)} ({backup_files[0][1]} MB)")

# 2. 로컬 메인 DB 복구
try:
    shutil.copy2(source_file, main_db_path)
    print(f"✅ 로컬 DB 복구 완료: {main_db_path}")
    
    # 복구된 파일 검증
    import duckdb
    con = duckdb.connect(database=main_db_path, read_only=True)
    tables = con.execute("SHOW TABLES").fetchall()
    print(f"✅ 복구 검증 성공: {len(tables)}개 테이블")
    
    # 테이블별 데이터 개수 확인
    for table in tables:
        table_name = table[0]
        try:
            count = con.execute(f"SELECT COUNT(*) FROM {table_name}").fetchone()[0]
            print(f"  📊 {table_name}: {count:,}개 레코드")
        except Exception as e:
            print(f"  ❌ {table_name}: 조회 실패 ({e})")
    
    con.close()
    
except Exception as e:
    print(f"❌ 로컬 DB 복구 실패: {e}")

# 3. Airflow 컨테이너용 디렉토리 생성 및 복사
try:
    # 디렉토리 생성
    os.makedirs('/data/duckdb', exist_ok=True)
    
    # 파일 복사
    shutil.copy2(main_db_path, airflow_db_path)
    print(f"✅ Airflow DB 복구 완료: {airflow_db_path}")
    
    # 파일 권한 설정 (Airflow에서 접근 가능하도록)
    os.chmod(airflow_db_path, 0o666)
    print("✅ 파일 권한 설정 완료")
    
except Exception as e:
    print(f"❌ Airflow DB 복구 실패: {e}")

# 4. 오래된 백업 파일들 정리 (크기가 작은 빈 파일들)
print("\n🧹 오래된 백업 파일 정리...")
small_backup_files = [
    '/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_044735',
    '/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_140542',
    '/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_143027',
    '/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_143605'
]

for backup_file in small_backup_files:
    try:
        if os.path.exists(backup_file):
            size_mb = os.path.getsize(backup_file) / (1024 * 1024)
            if size_mb < 0.1:  # 100KB 미만 파일 삭제
                os.remove(backup_file)
                print(f"🗑️ 삭제됨: {os.path.basename(backup_file)} ({size_mb:.2f} MB)")
    except Exception as e:
        print(f"❌ 삭제 실패 {backup_file}: {e}")

print("\n✅ 데이터베이스 복구 작업 완료!")
```

```python
# WAL 파일 문제 해결 및 깨끗한 DB 재생성
import os
import duckdb

print("🔧 WAL 파일 문제 해결...")

# 1. WAL 파일 삭제
wal_file = '/home/grey1/stock-kafka3/data/duckdb/stock_data.db.wal'
if os.path.exists(wal_file):
    os.remove(wal_file)
    print("🗑️ WAL 파일 삭제 완료")

# 2. 기존 메인 DB 파일 삭제
main_db_path = '/home/grey1/stock-kafka3/data/duckdb/stock_data.db'
if os.path.exists(main_db_path):
    os.remove(main_db_path)
    print("🗑️ 기존 메인 DB 파일 삭제")

# 3. 백업에서 새로운 DB 생성 (WAL 없이)
source_file = '/home/grey1/stock-kafka3/data/duckdb/stock_data.db.backup_20250727_044639'

try:
    # 소스 파일 연결 및 모든 데이터 추출
    source_con = duckdb.connect(database=source_file, read_only=True)
    
    # 새 DB 파일 생성
    new_con = duckdb.connect(database=main_db_path)
    
    # 테이블 목록 가져오기
    tables = source_con.execute("SHOW TABLES").fetchall()
    print(f"📊 복사할 테이블 수: {len(tables)}")
    
    for table in tables:
        table_name = table[0]
        print(f"  📋 복사 중: {table_name}")
        
        # 테이블 스키마 가져오기
        schema_info = source_con.execute(f"PRAGMA table_info('{table_name}')").fetchall()
        
        # CREATE TABLE 문 생성
        columns = []
        for col in schema_info:
            col_name, col_type, not_null, default_value, pk = col[1], col[2], col[3], col[4], col[5]
            col_def = f"{col_name} {col_type}"
            if not_null:
                col_def += " NOT NULL"
            if pk:
                col_def += " PRIMARY KEY"
            columns.append(col_def)
        
        create_sql = f"CREATE TABLE {table_name} ({', '.join(columns)})"
        new_con.execute(create_sql)
        
        # 데이터 복사
        data = source_con.execute(f"SELECT * FROM {table_name}").fetchall()
        if data:
            placeholders = ', '.join(['?' for _ in range(len(schema_info))])
            insert_sql = f"INSERT INTO {table_name} VALUES ({placeholders})"
            new_con.executemany(insert_sql, data)
        
        count = len(data)
        print(f"    ✅ {count:,}개 레코드 복사 완료")
    
    source_con.close()
    new_con.close()
    
    print("✅ 깨끗한 DB 파일 생성 완료")
    
    # 4. Airflow 경로로 복사
    airflow_db_path = '/data/duckdb/stock_data.db'
    if os.path.exists(airflow_db_path):
        os.remove(airflow_db_path)
    
    shutil.copy2(main_db_path, airflow_db_path)
    os.chmod(airflow_db_path, 0o666)
    print("✅ Airflow DB 복사 완료")
    
    # 5. 검증
    verify_con = duckdb.connect(database=main_db_path, read_only=True)
    tables = verify_con.execute("SHOW TABLES").fetchall()
    print(f"\n🔍 검증 결과: {len(tables)}개 테이블")
    
    for table in tables:
        table_name = table[0]
        count = verify_con.execute(f"SELECT COUNT(*) FROM {table_name}").fetchone()[0]
        print(f"  📊 {table_name}: {count:,}개 레코드")
    
    verify_con.close()
    
except Exception as e:
    print(f"❌ DB 재생성 실패: {e}")

print("\n✅ 모든 복구 작업 완료!")
```
